```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# Load the iris dataset
iris = datasets.load_iris()



# Prepare the data
X = iris.data
y = iris.target

# Build the K-Means model and fit the data
kmeans = KMeans(n_clusters=3).fit(X)

# Build the GMM model and fit the data
gmm = GaussianMixture(n_components=3).fit(X)

# Plot the results
colormap = np.array(['red', 'lime', 'black'])
plt.figure(figsize=(14,14))

plt.subplot(2, 2, 1)
plt.scatter(X[:, 2], X[:, 3], c=colormap[y], s=40)
plt.title('Real Clusters')

plt.subplot(2, 2, 2)
plt.scatter(X[:, 2], X[:, 3], c=colormap[kmeans.labels_], s=40)
plt.title('K-Means Clustering')

plt.subplot(2, 2, 3)
gmm_y = gmm.predict(X)
plt.scatter(X[:, 2], X[:, 3], c=colormap[gmm_y], s=40)
plt.title('GMM Clustering')

# Print a statement
print('Observation: GMM clustering matched the true labels more closely than K-Means.')
```
